动物繁殖模型，各年龄段为
$[x_1(k),x_2(k),\cdots,x_n(k)]$其中$k$为时间，
$s_i$为成活率，$b_i$为繁殖率

$$x_{1}(k+1)=\sum_{i=1}^nx_i(k)\\
x_{i+1}(k+1)=s_ix_{i}(k)
$$

因为$x_n$无法繁殖，故$i\ne n$，写成矩阵形式

$$L=\begin{bmatrix}b_1&b_2&\cdots&b_{n-1}&b_{n}\\
s_1&0&\cdots&0&0\\
0&s_2&\cdots&0&0\\
\vdots& \vdots&\vdots&\vdots&\vdots\\
0&0&\cdots&s_{n-1}&0\\
\end{bmatrix}\\X^T(k)=X^T(k-1)L 
$$


```python
X=np.full(5,100).T
L=np.zeros((5,5))
L[0,:]=np.array([0,0.2,1.8,0.8,0.2])
L[1,0]=0.5
L[2,1]=0.8
L[3,2]=0.8
L[4,3]=0.1
L
```

```python
num=30

arr=np.array([X])

for i in range(num):
    X=np.matmul(L,X)#先升维再相加
    arr=np.append(arr,[X],axis=0)
D=pd.DataFrame(arr.T)
D.index=['x1','x2','x3','x4','x5']
D
```

```python
D_p=D.copy()
for i in range(num+1):#还有初始数据
    D_p[i]=D[i]/D[i].sum()
D_p
#归一化处理
```
