```python
import pandas as pd
import numpy as np
import cufflinks as cf
import plotly.graph_objects as go

#SIR model
N0=0.99
NI0=0.01
v=0.3
R0=2
r=1
beta=v*R0
times=2000
dt=0.03
S=[N0]
I=[NI0]
R=[0]

T=range(times)
for i in T:
    S.append(S[i]-(beta*S[i]*I[i])*dt)
    I.append(I[i]+(beta*S[i]*I[i]-v*I[i])*dt)
    R.append(R[i]+v*I[i]*dt)

fig=go.Figure()
fig.add_trace(go.Scatter(x=np.array(T),y=S,name='S')) 
fig.add_trace(go.Scatter(x=np.array(T),y=I,name='I')) 
fig.add_trace(go.Scatter(x=np.array(T),y=R,name='R')) 
```

```python
#SIRS model
N0=0.99
NI0=0.01
v=0.3
r=0.5
beta=v*R0
d=0.2
times=1000
dt=0.1
S=[N0]
I=[NI0]
R=[0]

T=range(times)
for i in T:
    S.append(S[i]+(-beta*S[i]*I[i]+r*R[i])*dt)
    I.append(I[i]+(beta*S[i]*I[i]-v*I[i])*dt)
    R.append(R[i]+(v*I[i]-r*R[i]-d*I[i])*dt)

fig=go.Figure()
fig.add_trace(go.Scatter(x=np.array(T),y=S,name='S')) 
fig.add_trace(go.Scatter(x=np.array(T),y=I,name='I')) 
fig.add_trace(go.Scatter(x=np.array(T),y=R,name='R')) 
```

```python
R_list=[0.5,1,1.5,2,3,4]
fig=go.Figure()

for R0 in R_list:
    beta=v*R0
    S=[N0]
    I=[NI0]
    R=[0]

    T=range(times)
    for i in T:
        S.append(S[i]+(-beta*S[i]*I[i]+r*R[i])*dt)
        I.append(I[i]+(beta*S[i]*I[i]-v*I[i])*dt)
        R.append(R[i]+(v*I[i]-r*R[i]-d*I[i])*dt)

    fig.add_trace(go.Scatter(x=np.array(T),y=I,name=f'R0={R0}'))
fig.update_layout(title="R0")
fig.show()
```
