```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import cufflinks as cf

data=pd.read_csv("data/university_data.csv")
data
```

```python
J=len(data.columns[3:])
I=len(data)
X=np.zeros((I,J))
for j in range(J):
    X[:,j]=data[data.columns[3:][j]]
X
```

```python
#去量纲
for j in range(J):
    X[:,j]/=np.linalg.norm(X[:,j])
X
```

```python
W=np.array([0.3,0.3,0.3,0.025,0.075])
V=np.zeros((I,J))
for i in range(I):
    V[i,:]=X[i,:]*W
V
```

```python
#寻找最优和最劣解
A_best=np.zeros(J)
A_worst=np.zeros(J)
for j in range(J):
    A_best[j]=max(V[:,j])
    A_worst[j]=min(V[:,j])
A_best,A_worst
```

```python
#计算距离
def get_dist(A,B):
    return np.sqrt((A-B)@(A-B))

ans=np.zeros(I)
for i in range(I):
    S_best=get_dist(V[i],A_best)
    S_worst=get_dist(V[i],A_worst)
    ans[i]=S_worst/(S_best+S_worst)
ans
```

```python
ans_index = sorted(range(len(ans)), key=lambda k: ans[k], reverse=True)
ans_index=[i+1 for i in ans_index]
ans_index
```
