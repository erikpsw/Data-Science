```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the cleaned dataset
data = pd.read_csv('cleanData.csv')

# Drop the response variable 'Win_Loss' and standardize the features
features = data.drop(['Win_Loss'], axis=1)
scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

# Using the Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(standardized_features)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# From the elbow plot, determine the optimal number of clusters
optimal_clusters = 3  # example value, replace with the actual optimal number

# Applying K-Means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(standardized_features)

# Adding the cluster labels to the original data
data['Cluster'] = clusters

# Generate a placeholder list of 302 opponents
full_opponents_list = [f'Opponent{i+1}' for i in range(302)]

# Verify the length of the opponents list matches the data
num_rows = len(data)
print(f"Number of rows in dataset: {num_rows}")
print(f"Number of opponents provided: {len(full_opponents_list)}")

# Ensure the list has 302 entries
if len(full_opponents_list) != num_rows:
    raise ValueError("Length of opponents list must match the number of rows in the data. Please ensure you have exactly 302 opponents listed.")

# Add the opponents to the dataframe
data['Opponent'] = full_opponents_list

# Display the first few rows to ensure the mapping is correct
print(data.head())

# Analyze cluster-related insights for each opponent
opponent_clusters = data.groupby('Opponent')['Cluster'].value_counts().unstack().fillna(0)
print(opponent_clusters)

# Save the results for further analysis or use
data.to_csv('data_with_opponents_and_clusters.csv', index=False)
opponent_clusters.to_csv('opponent_clusters.csv')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the cleaned dataset
data = pd.read_csv('cleanData.csv')

# Display the first few rows of the dataset
print(data.head())

# Display the column names to identify the correct names
print(data.columns)

# Drop the response variable 'Win_Loss'
features = data.drop(['Win_Loss'], axis=1)

# Standardizing the features
scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

# Using the Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(standardized_features)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# From the elbow plot, determine the optimal number of clusters
optimal_clusters = 3  # example value, replace with the actual optimal number

# Applying K-Means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(standardized_features)

# Adding the cluster labels to the original data
data['Cluster'] = clusters

# Compute the average features for each cluster
cluster_averages = data.groupby('Cluster').mean()

# Display the cluster averages
print(cluster_averages)

# Save the cluster averages for further use
cluster_averages.to_csv('cluster_averages.csv')

# Example list of opponents for the 2024 season
opponents_2024 = ['Opponent1', 'Opponent2', 'Opponent3']  # Replace with actual opponents

# Adding 'Opponent' column for demonstration (since it doesn't exist in the provided columns)
# This is just an example, you should replace this with actual opponent data
data['Opponent'] = np.random.choice(opponents_2024, len(data))

# Mapping clusters to opponents
opponent_clusters = {}
for opponent in opponents_2024:
    opponent_data = data[data['Opponent'] == opponent]  # Filter data for the specific opponent
    if not opponent_data.empty:
        most_common_cluster = opponent_data['Cluster'].mode()[0]  # Find the most common cluster
        opponent_clusters[opponent] = most_common_cluster

# Display the mapping of opponents to clusters
print(opponent_clusters)

# Compute the average features for each opponent's cluster
opponent_averages = {opponent: cluster_averages.loc[cluster] for opponent, cluster in opponent_clusters.items()}

# Display the opponent averages
print(opponent_averages)

# Save the opponent averages for further use
opponent_averages_df = pd.DataFrame(opponent_averages).T
opponent_averages_df.to_csv('opponent_averages.csv')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
data = pd.read_csv('cleanData.csv')

# Drop the response variable 'Win_Loss' and standardize the features
features = data.drop(['Win_Loss'], axis=1)
scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

# Applying K-Means with the optimal number of clusters
optimal_clusters = 3  # Assuming 3 clusters based on previous analysis
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(standardized_features)

# Adding the cluster labels to the original data
data['Cluster'] = clusters

# Apply PCA to reduce dimensions to 2D
pca = PCA(n_components=2)
principal_components = pca.fit_transform(standardized_features)
data['PCA1'] = principal_components[:, 0]
data['PCA2'] = principal_components[:, 1]

# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=data, palette='viridis', s=100, alpha=0.6)
plt.title('Clusters Visualization using PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

Step 1: Apply PCA to reduce dimensions
First, we'll reduce the dimensions of the data to 2D using PCA.
Step 2: Plot the clusters
Next, we'll plot the clusters in the reduced 2D space.Each point represents a game, reduced to two principal components (PCA1 and PCA2).
The points are colored based on their cluster assignment.

```python
# The visualizations above show:

# Mean of Each Feature for Each Cluster:

# This bar plot displays the mean values of each feature across the different clusters.
# It helps identify the average characteristics of each cluster.
# Standard Deviation of Each Feature for Each Cluster:

# This bar plot shows the standard deviation values of each feature across the different clusters.
# It helps understand the variability within each cluster.

import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
data = pd.read_csv('cleanData.csv')

# Apply the previously determined clusters to the dataset
data['Cluster'] = clusters

# Calculate mean and standard deviation for each feature in each cluster
cluster_means = data.groupby('Cluster').mean()
cluster_stds = data.groupby('Cluster').std()

# Plotting the mean of each feature for each clusterP
plt.figure(figsize=(14, 8))
cluster_means.plot(kind='bar', figsize=(15, 8))
plt.title('Mean of Each Feature for Each Cluster')
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.legend(title='Cluster')
plt.show()

# Plotting the standard deviation of each feature for each cluster
plt.figure(figsize=(14, 8))
cluster_stds.plot(kind='bar', figsize=(15, 8))
plt.title('Standard Deviation of Each Feature for Each Cluster')
plt.xlabel('Features')
plt.ylabel('Standard Deviation')
plt.legend(title='Cluster')
plt.show()
```

```python
# The histograms for each feature within each cluster have been generated. These histograms provide a detailed view of the distribution of each feature within the different clusters.

# Each histogram shows:

# The frequency distribution of the feature values within each cluster.
# Distinct colors representing different clusters.


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
data = pd.read_csv('cleanData.csv')

# Apply the previously determined clusters to the dataset
data['Cluster'] = clusters

# Get the list of features (excluding the 'Cluster' and 'Win_Loss' columns)
features = data.columns.drop(['Cluster', 'Win_Loss'])

# Plot histograms for each feature within each cluster
for feature in features:
    plt.figure(figsize=(10, 6))
    for cluster in data['Cluster'].unique():
        sns.histplot(data[data['Cluster'] == cluster][feature], kde=False, label=f'Cluster {cluster}', alpha=0.6, bins=15)
    plt.title(f'Histogram of {feature} by Cluster')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```
