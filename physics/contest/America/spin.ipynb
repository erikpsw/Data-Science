```python
import numpy as np
from scipy.spatial.transform import Rotation
import copy

theta=np.radians(30)
v=10
vel=np.array([v*np.cos(theta),0,v*np.sin(theta)])

N=1000
t_max=10
dt=t_max/N
G=6.67e-11
m=190
M=5.965e24
R=6371.393e3
# h=400e3
h=0
pos=np.array([0.,0.,h])
pos_list=[]

for i in range(N):
    tmp=np.cross(vel,(np.cross(vel,np.array([0,0,1]))))
    axis=tmp*dt/np.linalg.norm(tmp)
    rotation = Rotation.from_rotvec(axis)
    vel=rotation.apply(vel)
    print(pos)
    d=1.5*np.e**(-h/6450)
    a=G*M*(1-d*v**2/(47**2))/(h+R)**2
    vel[2]-=a*dt
    pos+=vel*dt
    pos_list.append(copy.deepcopy(pos))
    
import plotly.graph_objects as go
data=np.array(pos_list)
fig = go.Figure(data=[go.Scatter3d(
    x=data[:,0],  # 3
    y=data[:,1], 
    z=data[:,2],
    mode='lines')])  # 'lines', 'markers', 'text'  3
fig.show()
```
