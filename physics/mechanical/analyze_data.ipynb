# FFT

对原始数据进行去噪、归一化和信号调整，确保数据的质量

```python
from scipy.fft import fft, fftfreq
# 数据预处理
def preprocess_signal(signal):
    # 去除直流分量
    signal = signal - np.mean(signal)
    # 归一化
    signal = signal / np.max(np.abs(signal))
    return signal

y = preprocess_signal(y)

# 计算采样频率
sampling_rate = 1 / (x[1] - x[0])

# 计算 FFT
N = len(x)
yf = fft(y) / N
xf = fftfreq(N, 1 / sampling_rate)

# 绘制频谱图
plt.figure(figsize=(10, 6), dpi=130)
plt.plot(xf, np.abs(yf), label='FFT Magnitude')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Spectrum')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 转化为单边
xf = xf[:N // 2]
yf = yf[:N // 2]
data = {
    "f_z": xf,
    "Magnitude": np.abs(yf)
}

# 创建 DataFrame
df = pd.DataFrame(data)

# 按照 Magnitude 降序排序
df_sorted = df.sort_values(by="Magnitude", ascending=False)

# 提取前几个强度最大的点及其对应的频率
top_n = 3  # 选择前几个点
top_frequencies = df_sorted.head(top_n)

top_frequencies
```

得到$f_n=1050\text{Hz}$

用计算出的频谱 对于激励信号在1点产生的频谱

```python
uff_file = pyuff.UFF('2251912潘世维/FFT(R)32_-X(f).uff')
data = uff_file.read_sets()
# print(data)
# 提取 x 和 data 数组
x = data['x']
y = data['data']

# 计算幅度
magnitude = np.abs(y)

# 可视化数据
plt.figure(figsize=(10, 6), dpi=130)
plt.plot(x, magnitude, label='FFT Magnitude')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
# 对数坐标
plt.yscale('log')
plt.title('Frequency Response')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = {
    "f_z": x,
    "Magnitude": magnitude
}

# 创建 DataFrame
df = pd.DataFrame(data)

# 按照 Magnitude 降序排序
df_sorted = df.sort_values(by="Magnitude", ascending=False)

# 提取前几个强度最大的点及其对应的频率
top_n = 3  # 选择前几个点
top_frequencies = df_sorted.head(top_n)

top_frequencies
```

同样也是$1050\text{Hz}$，验证成功，下面计算阻尼比，下降3dB，则

$$10\lg\frac 12=-3\text{dB}$$

```python
# Damping ratio calculation
peak_freq = 1050
half_power_point = 0.243463 / np.sqrt(2)
peak_index = 840
left_idx = np.where(magnitude[:peak_index] < half_power_point)[0][-1]
right_idx = np.where(magnitude[peak_index:] < half_power_point)[0][0] + peak_index
f1 = x[left_idx]  # Changed from xf to x
f2 = x[right_idx]  # Changed from xf to x
damping_ratio = (f2 - f1) / (2 * peak_freq)

offset = 200
plt.figure(figsize=(10, 6), dpi=130)
plt.axhline(y=half_power_point, color='r', linestyle='--', label='Half Power Points')
plt.plot(df["f_z"][left_idx], magnitude[left_idx], 'bo', label='Left Index Point')
plt.plot(df["f_z"][right_idx], magnitude[right_idx], 'go', label='Right Index Point')
plt.plot(df["f_z"][left_idx-offset:right_idx+offset],magnitude[left_idx-offset:right_idx+offset], label='local Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Damping Ratio Visualization with Indices')
plt.legend()
plt.grid(True)
plt.show()

print("阻尼比",damping_ratio)
```
