```python
import pyuff
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
rcParams['font.family'] = 'Times New Roman'  # 设置字体
uff_file = pyuff.UFF('2251912潘世维/Coh(E_R)1_-X,32_-X(f).uff')
data = uff_file.read_sets()
data
```

```python
# 提取 x 和 data 数组
x = data['x']
y = data['data']

# 可视化数据
plt.figure(figsize=(10, 6), dpi=130)
plt.plot(x, y, label='Coh(E_R)1_-X,32_-X(f)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Response')
plt.legend()
plt.grid(True)
plt.show()
```

```python
uff_file = pyuff.UFF('2251912潘世维/FFT(E)1_-X(f).uff')
data = uff_file.read_sets()
print(data)
# 提取 x 和 data 数组
x = data['x']
y = data['data']

# 计算幅度
magnitude = np.abs(y)

# 可视化数据
plt.figure(figsize=(10, 6), dpi=130)
plt.plot(x, magnitude, label='FFT Magnitude')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.yscale('log')
plt.title('Frequency Response')
plt.legend()
plt.grid(True)
plt.show()
```
