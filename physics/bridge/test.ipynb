```python
import numpy as np
import pandas as pd
import math
import os

import plotly.graph_objects as go
Y_arr=[]
t_arr=[]
fig=go.Figure()
data_file = 'data.csv'
F=[]
X=[]

F2=[]
X2=[]
with open(data_file, 'r',encoding='utf-8', errors='ignore') as f:
    for line in f:
        x=float(line.split(',')[1])
        F.append(float(line.split(',')[0]))
        X.append(x)
        if(x>0.3365 and x<3.1924):
            F2.append(float(line.split(',')[0]))
            X2.append(x)
        
fig.add_trace(go.Scatter(x=X,y=F,name='实验数据'))
    

fig.update_layout(title='三点弯加载曲线  制作者:潘世维',
                  xaxis_title='x/mm',yaxis_title='F/N')
```

```python
import statsmodels.api as sm
import numpy as np

x = sm.add_constant(X2) # 若模型中有截距，必须有这一步
model = sm.OLS(F2, X2).fit() # 构建最小二乘模型并拟合
print(model.summary()) # 输出回归结果

 
```

```python
predicts = model.predict() # 模型的预测值

fig=go.Figure()
fig.add_trace(go.Scatter(x=X2,y=F2,name='实验数据'))
fig.add_trace(go.Scatter(x=X2,y=predicts,name='拟合直线',line ={'dash':'dot'}))
fig.update_layout(title='三点弯加载曲线  制作者:潘世维',
                  xaxis_title='x/mm',yaxis_title='F/N')
```
