```python
import numpy as np
import pandas as pd
import math
import os

import plotly.graph_objects as go
Y_arr=[]
t_arr=[]
fig=go.Figure()
data_file = '1.csv'
time=[]
Y=[]
with open(data_file, 'r') as f:
    for line in f:
        time.append(float(line.split('	')[0])-5.997106000E-1)
        Y.append(float(line.split('	')[1])+1.404)
    # Y=np.flipud((np.array(Y)-Y[-1])/100)[::-1]
    Y_arr.append(Y)
    t_arr.append(time)
    fig.add_trace(go.Scatter(x=time,y=Y,name='实验数据'))
    
g=-9.8 #重力加速度
dt=0.0005#时间间隔
n=6100#模拟次数
t=0#初始时刻
R=0.02#小球半径
e=0.79678#恢复系数
y=1.404#初始时的y坐标
vy=0 #y方向的速度
count=0
T=[t]
Y=[y]
#主循环
for i in range(n):
    if(y-R<0):
        count+=1
        if(count==1):
            vy=-e*vy
    else:
        count=0
    t=t+dt
    y=y+vy*dt#更新位置
    vy=vy+g*dt#更新速度
    T.append(t)
    Y.append(y)
fig.add_trace(go.Scatter(x=T,y=Y,name='理论数据')) 
fig.update_layout(title='小球的y坐标与时间的关系  制作者:潘世维',
                  xaxis_title='t/s',yaxis_title='y/m')
```
