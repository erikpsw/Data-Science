```python
import plotly.graph_objects as go
g=-9.8 #
dt=0.0005#
n=8000#
t=0#
R=0.05#
e=0.8#
y=1#y
vy=0 #y
count=0
T=[t]
Y=[y]
#
for i in range(n):
    if(y-R<0):
        count+=1
        if(count==1):
            vy=-e*vy
    else:
        count=0
    t=t+dt
    y=y+vy*dt#
    vy=vy+g*dt#
    T.append(t)
    Y.append(y)

fig=go.Figure()
fig.add_trace(go.Scatter(x=T,y=Y)) 
fig.update_layout(title='y  :',
                  xaxis_title='t/s',yaxis_title='y/m')
```
