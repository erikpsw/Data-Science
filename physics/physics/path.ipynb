```python
import plotly.graph_objects as go
g=-9.8 #重力加速度
dt=0.0005#时间间隔
n=8000#模拟次数
t=0#初始时刻
R=0.05#小球半径
e=0.8#恢复系数
y=1#初始时的y坐标
vy=0 #y方向的速度
count=0
T=[t]
Y=[y]
#主循环
for i in range(n):
    if(y-R<0):
        count+=1
        if(count==1):
            vy=-e*vy
    else:
        count=0
    t=t+dt
    y=y+vy*dt#更新位置
    vy=vy+g*dt#更新速度
    T.append(t)
    Y.append(y)

fig=go.Figure()
fig.add_trace(go.Scatter(x=T,y=Y)) 
fig.update_layout(title='小球的y坐标与时间的关系  制作者:潘世维',
                  xaxis_title='t/s',yaxis_title='y/m')
```
