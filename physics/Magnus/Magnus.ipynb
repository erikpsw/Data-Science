```python
import plotly.graph_objects as go
X_arr=[]
Y_arr=[]
t_arr=[]
fig=go.Figure()
data_file = os.path.join('data', f'01.csv')
time=[]
X=[]
Y=[]
with open(data_file, 'r') as f:
    for line in f:
        time.append(float(line.split('	')[0]))
        X.append(float(line.split('	')[1]))
        Y.append(float(line.split('	')[2]))
    X=np.flipud((np.array(X)-X[-1])/100)
    Y=np.flipud((np.array(Y)-Y[-1])/100)[::-1]
    X_arr.append(X)
    Y_arr.append(Y)
    t_arr.append(time)
    fig.add_trace(go.Scatter(x=X,y=Y,name=f'实验数据'))
  
#分析角度 到0为止
g=-9.8
n=300
dt=0.01
vx=2
vy=8
x=0
y=3.3
m=1
def update3(x=0,y=2.553,vx=1.533,vy=1.66,l=0.001,l2=0.0033):
    #x,y为初始点的坐标，vx,vy分别为x,y方向速度的大小,l为阻力系数,l2为升力系数
    t=0
    X=[x]
    Y=[y]
    for i in range(n):
        t=t+dt
        x=x+vx*dt
        y=y+vy*dt
        v=math.sqrt(vx**2+vy**2)
        omega=10*math.e**(-0.5*t)
        aF=omega*v*l2
        af=l*v**2/m
        vx=vx-af*(vx/v)-aF*(vy/v)
        vy=vy+g*dt-af*(vy/v)+aF*(vx/v)
        X.append(x)
        Y.append(y)
        if(y<0):
            break
    return X,Y
X3,Y3=update3()
fig.add_trace(go.Scatter(x=X3,y=Y3,name=f'理论计算'))
fig.update_layout(title='马格努斯飞行器理论轨迹分析  制作者:潘世维',
                    xaxis_title='x',yaxis_title='y')
```

```python
fig=go.Figure()
for i in range(30,80,5):
     tan=math.tan(i*math.pi/180)
     X3,Y3=update3(x=0,y=2.553,vx=3,vy=vx*tan,l=0.001,l2=0.001)
     fig.add_trace(go.Scatter(x=X3,y=Y3,name=f'角度{i}'))
     
fig.update_layout(title='马格努斯飞行器理论轨迹分析  制作者:潘世维',
               xaxis_title='x',yaxis_title='y')
```
