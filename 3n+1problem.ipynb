```python
def result(n):
    steps=[]
    while n!=1:
        if n%2==0:
            n=n/2
        else:
            n=3*n+1
        steps.append(n)
    return(steps)
len(result(27))
```

```python
import pandas as pd

S=pd.Series(result(27))
S.plot()
```

```python
max(result(27))
```

```python
res=[]
cou=[]
for i in range(30,100):
    ste=[i]
    cur=i
    count=0
    while cur!=1:
        if cur%2==0:
            cur=cur/2
        else:
            cur=3*cur+1
        ste.append(cur)
        count=count+1
    res.append(max(ste))
    cou.append(count)
    


pd.Series(res).plot()  
```

```python
pd.Series(cou).plot() 
```
