```python
import numpy as np
import cufflinks as cf
import plotly.graph_objects as go

def impulse(x):return np.where(x == 0, 1, 0)
def step(x):return np.where(x >= 0, 1, 0)

class sig():
    def __init__(self,exp) -> None:
        self.exp=exp
        self.X=0
        self.Y=0
        
    def setX(self,X):
        self.X=np.array(X)
        self.Y=self.exp(self.X)
    
    def graph(self,mode=0):

        if(type(self.Y[0])==complex or np.complex128):
            fig = go.Figure().set_subplots(2,1,
    subplot_titles=("实部","虚部"))
            if(mode==1):
                fig.add_trace(go.Scatter(x=self.X,y=self.Y[:].real,mode="markers"))
                fig.add_trace(go.Scatter(x=self.X,y=self.Y[:].imag,mode="markers"),row=2,col=1)
            else:
                fig.add_trace(go.Scatter(x=self.X,y=self.Y[:].real))
                fig.add_trace(go.Scatter(x=self.X,y=self.Y[:].imag),row=2,col=1)
            fig.show()
        else:
            fig=go.Figure()
            if(mode==1):
                fig.add_trace(go.Scatter(x=self.X,y=self.Y),mode="markers")
            else:
                fig.add_trace(go.Scatter(x=self.X,y=self.Y))
            fig.show()
        
    def __add__(self, other):
        return sig(lambda X:self.exp(X)+other.exp(X))
    
    def __sub__(self, other):
        return sig(lambda X:self.exp(X)-other.exp(X))

    
sig1=sig(step)-sig(lambda x:(0.5+3j)**x)
sig1.setX(np.linspace(0,5,100))
sig1.graph()
```
