$np.random.randint\,\,[Min,Max)$

```python
s1=pd.Series(100)
i=0
while(i<100):
    t=sum(np.random.randint(35,46,size=5)/50)
    if(t>3.5 and t<4.5):
        s1[i]=t
        i=i+1
    else:
        pass
s1.plot()
```

```python
sns.distplot(s1,hist=True,kde=True,bins=20)
```
