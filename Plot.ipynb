```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
columns=list('ABCD'),
index=np.arange(0, 100, 10))
print(df)
df.plot()
```

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(color='k', alpha=0.7)
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
index=['one', 'two', 'three', 'four', 'five', 'six'],
columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
#使用了列索引，故颜色不同
df.plot.bar()
```

```python
df.plot.barh(stacked=True, alpha=0.5)#提高空间利用率
```

```python
comp1 = np.random.normal(0, 1, size=200)#正态分布
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values,hist=True,kde=True,bins=100)#hist是直方图，kde是核密度，bins是分成的区间数
```

```python
macro = pd.read_csv('csv/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]#对数差分，dropna滤除缺失数据
sns.regplot('m1', 'unemp', data=trans_data)#线性回归
```

```python
sns.pairplot(trans_data,diag_kind='kde',plot_kws={'alpha': 0.2})
#对角线上是分布图,diag_kind有"hist"与"kde"
#kind是非对角线，有"scatter"与"reg"
```
