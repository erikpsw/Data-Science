```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

n=100
A=[0.4]
R=3.1
for i in range(1,n-1):
    A.append(round(R*A[i-1]*(1-A[i-1]),3))
print(A)
```

```python
s = pd.Series(A)
s.plot()
```

```python
n=400
point = []
k=0
for j in range(1,n):
    A=[0.4]
    R=j/100
    for i in range(1,n-1):
        A.append(round(R*A[i-1]*(1-A[i-1]),3))
    b=set(A[80:])
    point.append(list(b))
print(point[:50])
```

```python
ax = plt.subplot(111)
for i in range(n-1):
    for j in point[i]:
        ax.scatter(i, j,marker = '.',s=3)
plt.show()
```
