```python
df_stores2 = df_stores.copy()

# 将 "dishs" 列转换为包含字典的列表
df_stores2["dishs"] = df_stores2.apply(lambda x: [{dish.split("_")[0]: int(dish.split("_")[1])} for dish in x["dishs"].split("|") if len(dish.split("_")) == 2], axis=1)

# 显示转换后的 DataFrame
df_stores2[["name", "dishs"]]
```
