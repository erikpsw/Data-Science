```python
import pandas as pd
import numpy as np
import cufflinks as cf

g=-9.8
n=300
dt=0.01
vx=2
vy=8
x=0
y=0
```

```python

#无空气阻力
def update1(x=0,y=0,vx=2,vy=8):
    X=[x]
    Y=[y]
    for i in range(n):
        x=x+vx*dt
        y=y+vy*dt
        vy=vy+g*dt
        X.append(x)
        Y.append(y)
    return X,Y
X,Y=update1(vx=2,vy=8)
```

```python
pd.DataFrame(Y).iplot()
```

```python
path=pd.DataFrame(Y,index=X)
path.iplot()
```

```python
m=1#有阻力
import math
def update2(x=0,y=0,vx=2,vy=8,l=0.001):
    X=[x]
    Y=[y]
    for i in range(n):
        x=x+vx*dt
        y=y+vy*dt
        
        v=math.sqrt(vx**2+vy**2)
        af=l*v**2/m
        vx=vx-af*(vx/v)
        vy=vy+g*dt-af*(vy/v)
        X.append(x)
        Y.append(y)
    return X,Y
X2,Y2=update2(vx=2,vy=8)
```

```python
#有升力有阻力
def update3(x=0,y=0,vx=2,vy=8,l=0.001,l2=0.001):
    t=0
    X=[x]
    Y=[y]
    for i in range(n):
        t=t+dt
        x=x+vx*dt
        y=y+vy*dt
        v=math.sqrt(vx**2+vy**2)
        omega=7*math.e**(-0.5*t)
        aF=omega*v*l2
        af=l*v**2/m
        vx=vx-af*(vx/v)-aF*(vy/v)
        vy=vy+g*dt-af*(vy/v)+aF*(vx/v)
        X.append(x)
        Y.append(y)
    return X,Y
X3,Y3=update3()
```

```python
#有升力
def update4(x=0,y=0,vx=2,vy=8,l2=0.001):
    t=0
    X=[x]
    Y=[y]
    for i in range(n):
        t=t+dt
        x=x+vx*dt
        y=y+vy*dt
        v=math.sqrt(vx**2+vy**2)
        omega=5*math.e**(-0.5*t)
        aF=omega*v*l2
        vx=vx-aF*(vy/v)
        vy=vy+g*dt+aF*(vx/v)
        X.append(x)
        Y.append(y)
    return X,Y
X4,Y4=update4()
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig=go.Figure()
fig.add_trace(go.Scatter(x=X,y=Y,name='无阻力升力')) 
fig.add_trace(go.Scatter(x=X2,y=Y2,name='阻力')) 
fig.add_trace(go.Scatter(x=X3,y=Y3,name='升力+阻力'))
fig.add_trace(go.Scatter(x=X4,y=Y4,name='升力'))
fig.update_layout(title='马格努斯飞行器理论轨迹分析  制作者:潘世维',
                  xaxis_title='x',yaxis_title='y')
```
