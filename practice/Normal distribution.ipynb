正态分布函数为

$p(x)=\frac{1}{\sigma \sqrt{2 \pi}} \cdot e^{-\frac{(x-\mu)^{2}}{2 \sigma^{2}}}$

期望值μ决定了其位置，其标准差σ决定了分布的幅度

其中标准差$\sigma$为$\sigma_{n}=\sqrt{\frac{\left(x_{1}-\bar{x}\right)^{2}+\left(x_{2}-\bar{x}\right)^{2}+\cdots+\left(x_{n}-\bar{x}\right)^{2}}{n}}$

```python
u = 0  # 均值μ
u01 = -2
sig = math.sqrt(0.2)  # 标准差δ

x = np.linspace(u - 3 * sig, u + 3 * sig, 50)
y_sig1 = np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (math.sqrt(2 * math.pi) * sig)

sig = math.sqrt(0.4)  

y_sig2 = np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (math.sqrt(2 * math.pi) * sig)

plt.plot(x, y_sig1, "r-", label=r"$\sqrt{0.2}$")
plt.plot(x, y_sig2, "b-",label=r"$\sqrt{2}$")
plt.legend(loc='upper right')#一定要加
plt.grid(True)
plt.show()
```

我们发现，方差越小越接近于均值

  $X_{1}$, $X_{2}$, $\ldots$, $X_{n}$  他们彼此之间满足独立同分布, 因此他们拥有相同 的均值  $\mu $ 和方差 $ \sigma^{2} $

我们有如下的随机变量：此时, 我们重点来研究这一组随机变量的均值: $ M_{n}=\frac{X_{1}+X_{2}+\ldots+X_{n}}{n}$

它的均值为

$E\left[M_{n}\right]=E\left[\frac{X_{1}+X_{2}+\ldots+X_{n}}{n}\right]=\frac{1}{n}\left(E\left[X_{1}\right]+E\left[X_{2}\right]+\ldots+E\left[X_{n}\right]\right)=\frac{1}{n} \cdot n \cdot \mu=\mu=E\left[X_{i}]\right.$

它的方差为

$\begin{array}{l}
\operatorname{var}\left[M_{n}\right]=\operatorname{var}\left[\frac{X_{1}+X_{2}+\ldots+X_{n}}{n}\right]=\frac{1}{n^{2}} \operatorname{var}\left[X_{1}+X_{2}+\ldots+X_{n}\right] \\
=\frac{1}{n^{2}}\left(\operatorname{var}\left[X_{1}\right]+\operatorname{var}\left[X_{2}\right]+\ldots+\operatorname{var}\left[X_{n}\right]\right)=\frac{1}{n^{2}} \cdot n \cdot \sigma^{2}=\frac{\sigma^{2}}{n}
\end{array}$

$\bar{X}_{n} \stackrel{d}{\rightarrow} N\left(\mu, \frac{\sigma^{2}}{n}\right)$

这就是中心极限定理
